```python
object_cols = df.select_dtypes(include='object').columns

dollar_cols = []
number_cols = []
percent_cols = []
date_cols = []

for col in object_cols:
    sample = str(df[col].iloc[0])

    if '$' in sample:
        dollar_cols.append(col)
    elif '%' in sample:
        percent_cols.append(col)
    elif ',' in sample and not any(c.isalpha() for c in sample.replace(',', '').replace('.', '').replace('-', '')):
        number_cols.append(col)
    elif sample.startswith('20') and '-' in sample:
        date_cols.append(col)

print("\n🗓️ คอลัมน์วันที่:")
for col in date_cols:
    print(f"  • {col}")

print("\n💵 คอลัมน์ราคา (มี $):")
for col in dollar_cols:
    print(f"  • {col}")

print("\n🔢 คอลัมน์ตัวเลข EU format:")
for col in number_cols:
    print(f"  • {col}")

print("\n📊 คอลัมน์เปอร์เซ็นต์:")
for col in percent_cols:
    print(f"  • {col}")
```

```python
date_cols = []
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')


dollar_cols = []
for col in dollar_cols:
    df[col] = df[col].astype(str).str.replace('$', '').str.replace(',', '').astype(float)


number_cols = []
for col in number_cols:
    df[col] = df[col].astype(str).str.replace('.', '').str.replace(',', '.').astype(float)


percent_cols = []
for col in percent_cols:
    df[col] = df[col].astype(str).str.replace('%', '').str.replace(',', '.').astype(float)

print("✅ Cleaning complete!")
```
